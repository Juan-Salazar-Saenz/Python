```python
import random
def moneda(N=10,r =['C','S']):
    if N == 0:
        raise Exception('NO PUEDE SER CERO!')
    if N < 1:
        raise Exception('NO PUEDE SER MENOR QUE 1')
    if type(N) != int:
        raise Exception('TIENE QUE SER NUMERO ENTERO!')
    lanzamientos = [] #almaceno cada lanzamiento
    for _ in range(N):
        lanza = random.choice(r) #lanzo la moneda
        lanzamientos.append(lanza) #Agrego el resultado a la lista
    return lanzamientos.count(r[0])/N, lanzamientos.count(r[1])/N 
print(moneda(3.2))
```

```python
#Defino una función de "limpieza"
import string
def limpieza(s):
  s = s.lower() #pongo el texto en minúsculas
  s = s.replace(' ','')
  #Se quitan las tildes
  #s = s.replace('á','a')
  #s = s.replace('é','e')
  #s = s.replace('í','i')
  #s = s.replace('ó','o')
  #s = s.replace('ú','u')
  for vocal_tilde, vocal in zip('áéíóú','aeiou'):
    s = s.replace(vocal_tilde,vocal)
  for p in string.punctuation + '¿¡':
    s = s.replace(p,'')
  return s

```

```python
def isPalindromo(s):
  "Se recibe la cadena sin espacios ni tildes y en minúscula"
  if s == s[::-1]:
    return True
  else:
    return False
```

```python
def palindromo1(s):
  if s.endswith(s[::-1]):
    print("Este texto se lee al derecho y al reves")
    return True
  else:
    print("Sigue intentando")
    return False
```

###Una función recursiva se llama así misma

##Necesita dos elementos:
##1. Cuándo parar
##2. Una iteración

```python
def factorial(N):
  if N ==1:
    return 1*N
  else:
    return N*factorial(N-1)
```

```python
def palin_recursivo(s):
  if len(s) <= 1:
    return True
  elif s[0] == s[-1]:
    return palin_recursivo(s[1:-1])
  else:
    return False
```
