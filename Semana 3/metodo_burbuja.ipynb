###MÉTODO DE LA BURBUJA

```python
lista = [64, 34, 25, 12, 22, 11, 90, 1, 1000, 40]
long_lista = len(lista)
#Se itera en cada elemento de la lista por índices
#Se recorre hasta el penúltimo elemento
for i in range(long_lista -1):
  for j in range(long_lista  -1):
    if lista[j] > lista[j + 1]:
      lista[j], lista[j + 1] = lista[j + 1], lista[j]
print(lista)
```
