###Construya una  función cuyas entradas sean:
1. Una lista con los posibles resultados de lanzar una moneda. Por ejemplo ['C','S']
2. Un número de lanzamientos. 

La función debe lanzar la moneda N veces, y al final la salida debería mostrar la fracción de caras y la fracción de sellos. 

¿Qué pasa si lanza la moneda 20, 50, 100, 500 o 1000 veces?

```python
random.random()
```

```python
def moneda(N=10,r =['C','S']):
  lanzamientos = [] #almaceno cada lanzamiento
  for _ in range(N):
    lanza = random.choice(r) #lanzo la moneda
    lanzamientos.append(lanza) #Agrego el resultado a la lista
  return lanzamientos.count('C')/N, lanzamientos.count('S')/N 

moneda(1000)
```

```python
import random
def monedaCargada(N,p=0.5):
  '''
  recibe un número N de lanzamientos
  con una probabilidad p de que sea cara
  y 1- p de que sea sello
  '''
  if p <0 or p>1:
    raise ValueError('p debe ser un número en [0,1]')
  lanzamientos = []
  r =['C','S']
  for _ in range(N):
    lanza = random.random()
    if lanza < p:
      lanzamientos.append('C')
    else:
      lanzamientos.append('S')
  return lanzamientos.count('C')/N, lanzamientos.count('S')/N
```

```python
monedaCargada(10000,1.1)
```

###Repita el ejercicio anterior, pero esta vez, para cada uno de los valores de N, lance la moneda dos veces, y agrupe los resultados en una lista tipo ['CS','CC',...]. 

###Al final de la rutina diga la fracción de cada resultado

```python
import random 
def dos_monedas(N):
  lanzamientos = []
  r = ['C','S']
  for _ in range(N):
    result = ''
    for i in range(2):
      lanz = random.choice(r)
      result += lanz
    lanzamientos.append(result)
  return lanzamientos

dos_monedas(10)
```
