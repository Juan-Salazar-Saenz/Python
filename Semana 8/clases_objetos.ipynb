###PROGRAMACIÓN ORIENTADA A OBJETOS

En programación, un objto es una estructura de datos personalizada que contiene datos y código.

image.png

Un objeto es un ente que interactúa a través de sus atributos y métodos.

Por ejemplo, si tomamos un automóvil como un objeto, algunos de sus atributos y métodos podrían ser:

Atributos:

*   Marca (tipo string)
*   Edad (tipo int)
*   Nuevo (tipo bool)
*   Placa (tipo string)


Métodos:

*   Llenar tanqueado 
*   Contar el kilometraje






Un objeto se crea a partir de una clase que lo contiene. En otras palabras, una clase es un "molde" para crear objetos, y un objeto será un ejemplo de dicha clase.

Veamos cómo se construiría una clase sencilla para el ejemplo del automóvil

```python
class Automovil:
  def __init__(self,marca,edad,nuevo=True):
    self.marca = marca
    self.edad = edad
    self.nuevo = True
    self.kms = 0 if True else None

  def contar_kms(self,recorrido):
    self.kms = self.kms + recorrido
    return self.kms
```

Note que la clase Auto tiene dos métodos (funciones)


*   __init__
*   contar_kms

El método ______init__ (con doble guion bajo antes y después) es el __contructor__, que es el método que va a almacenar los __atributos__ de la clase. 

Note que los atributos de la clase se almacenan con self.atributo, y que todos los métodos empiezan con self como primer argumento. LA palabra self reconocerá que cada atributo y método pertenece a la clase



```python
#para ver los métodos, dir es indicado
dir(Automovil)
```

```python
#creamos dos nuevos OBJETOS: un automóvil Chevrolet y un automóvil Mazda
chevrolet = Automovil('Chevrolet',10,False)
print(chevrolet.kms)
#print(chevrolet.edad) #Se llama el atributo edad del objeto chevrolet
#chevrolet.contar_kms(5000)
```

```python
chevrolet.contar_kms(1000)
chevrolet.kms
```

```python
mazda = Automovil('Mazda',0)
print(mazda.marca, mazda.nuevo,mazda.kms)
```

###EJEMPLO 2: CUENTA BANCARIA

Observe el ejemplo de una cuenta bancaria, donde los atributos serán el nombre del propietario, el número de cuenta, y su balance. 

Los métodos serán depositar, retirar y mostrar_info

```python
class Cuenta:
  #contructor para definir los atributos
  def __init__(self,nombre,num_cuenta,tipo = 'Ahorros',cantidad_inicial=0):
    self.nombre = nombre
    self.num_cuenta = num_cuenta
    self.tipo = tipo
    self.balance = cantidad_inicial

  #método para depositar dinero
  def depositar(self,cantidad):
    self.balance += cantidad

  def retirar(self,cantidad):
    self.balance -= cantidad

  def mostrar_info(self):
    info = 'el usuario {} con cuenta no. {} tiene un saldo de {}'.format(self.nombre,self.num_cuenta,self.balance)
    print(info)

  def mostrar_retirar(self):
    if self.balance <= 0:
      info1 = 'estimado {} usted no tiene saldo'.format(self.nombre,self.balance)  
      print(info1)
    else:
      self.mostrar_info()
```

```python
juan = Cuenta('Juan García','11110000',2000000)
ana = Cuenta('Ana Valencia','22220000',0)
```

```python
juan.mostrar_retirar()
```
