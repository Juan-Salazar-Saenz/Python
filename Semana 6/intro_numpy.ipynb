```python
def factorial(n):
  if n==1 or n==0: 
    return 1
  else:
    return n*factorial(n-1)
```

```python
%timeit [factorial(i) for i in range(50)]
```

```python
%%timeit
factoriales = []
for i in range(50):
  factoriales.append(factorial(i))
```

```python
def binomial(k,N,p=0.5):
  '''
  ¿Cuál es la probabilidad de obtener k de N elementos
  '''
  return math.factorial(N)/(math.factorial(k)*math.factorial(N-k))*p**N*(1-p)**(N-k)
```

```python
def gaussiana(x,mu=0,sigma=1):
  return 1/np.sqrt(2*np.pi*sigma**2)*np.exp(-(x-mu)**2/(2*sigma**2))
```

```python
x =np.linspace(-5,5,100)
y1 = gaussiana(x)
y2 = gaussiana(x,mu=1)
plt.plot(x,y1,label='$\mu=0$')
#plt.plot(x,y2,label= '$\mu=1$')
plt.fill_between(x,y2,y1,alpha=0.3,label='$mu=1$')
plt.title('Distribuciones gausssianas')
plt.yticks([0,0.2,0.4])
plt.xticks([-4,0,4])
plt.xlabel('x')
plt.ylabel('P(x)')
plt.legend()
plt.grid()
```
