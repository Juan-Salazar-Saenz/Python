###CICLOS FOR

```python
animal = 'perro'
for l in animal: 
  l = l.capitalize()
  print(l)
```

```python
import string 
puntuacion = string.punctuation
#print(puntuacion)
for punt in puntuacion:
  if punt == '!' or punt == '?':
    print(punt)
```

```python
numeros = [1,23,3,6,7,80,30,55]
for num in numeros: 
  if num%2 == 0:
    break  
  else:
    print(num)
```
