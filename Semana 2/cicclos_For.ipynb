```python
import string

texto = """"El coronel abrió el tarro del café y comprobó que no había más de una cucharadita. Retiró la olla del fogón,
   vertió la mitad del agua en el piso de tierra, y con un cuchillo raspó el interior del tarro
   sobre la olla hasta cuando se desprendieron las últimas raspaduras del polvo de café revueltas con óxido de lata...""""

# Crear una tabla de traducción para eliminar las tildes
tabla_tildes = str.maketrans('áéíóú', 'aeiou')

# Eliminar las tildes del texto
texto_sin_tildes = texto.translate(tabla_tildes)

# Eliminar los signos de puntuación del texto
texto_sin_tildes_ni_puntuacion = texto_sin_tildes.translate(str.maketrans('', '', string.punctuation))

print(texto_sin_tildes_ni_puntuacion)
```
