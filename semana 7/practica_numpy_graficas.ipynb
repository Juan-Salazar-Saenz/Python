###complete la siguiente función

```python
import numpy as np
def aleatorio(m,n):
  '''
  m: número de filas
  n: número de columnas
  regresa una matriz de tamaño (m,n+1)
  de números al azar con los valores de la última columna
  igual a 1'''
  numeros = np.random.randint(0,99,(m,n+1))
  numeros[:,-1] = 1
  return numeros

print(aleatorio(5,6))
print(aleatorio(5,6).T)
```

###Explore las funciones matemáticas de numpy exp, seno, coseno, tan, y escriba en una función lambda las distribuciones binomial, gauss y poisson

```python
import math
binomial = lambda x,n,p: math.comb(n,x)*(p**x)*(1-p)**(n-x)

binomial(4,4,1/6)
```

```python
binomiales = np.random.binomial(2000,0.005,(50,))
import matplotlib.pyplot as plt
plt.hist(binomiales,density=False)
```

```python
bin10 = binomiales[binomiales ==10]
len(bin10) 
```
