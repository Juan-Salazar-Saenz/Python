```python
import numpy as np
```

```python
#se crean dos arreglos (arrays)
a1 = np.arange(20).reshape(4,5)
a2 = np.random.randint(-10,10,size=(2,5))
a3 = np.arange(8).reshape(4,2)
print(a1)
print(a2)
print(a3)
```

```python
#¿Què ocurre cuando corre este comando?
np.vstack((a1,a2))
```

```python
#Corra los siguientes comandos
#np.hstack((a3,a1))
np.hstack((a1,a3))
```

```python
np.concatenate((a1,a3),axis=1)
```
