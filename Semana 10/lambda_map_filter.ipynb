```python
def intdiv(a,b):
  try:
    result = a//b
    return result
  except TypeError:
    print('Tipo de variable no permitida')
  except ZeroDivisionError:
    print('No se puede dividir entre cero')
  except Exception:
    print('algo salió mal')
```

```python
import string
def limpieza(s):
  s = s.lower().replace(' ','')
  for vocal_tilde,vocal in zip('áéíóú','aeiou'):
    s = s.replace(vocal_tilde,vocal)
  for punt in string.punctuation + '¿¡':
    s = s.replace(punt,'')
  return s 

def espalindromo(s):
  if len(s) == 1 or len(s) == 0:
    return True
  elif len(s) == 2:
    if s[0] == s[1]:
      return True
    else:
      return False
  else:
    if s[0] == s[-1]:
      return espalindromo(s[1:-1])
    else:
      return False
```

```python
s = ["in girum imus nocte et consumimur igni", 'Atar a la rata','¿Acaso hubo búhos acá?','¿Son robos o sobornos?','A ti no, bonita.']
for i in s:
  i = limpieza(i)
  print(i)
  print(espalindromo(i))
print(espalindromo('ana'))
```

```python
def P(p):
  if (not type(p) is float) and (not type(p) is int):
    raise TypeError(f'p debe ser un número entero o real')
  if p<0 or p>1:
    raise ValueError(f'p debe estar en el intervalo (0,1]')
  return p
```
