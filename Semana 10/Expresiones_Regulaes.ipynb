##EXPRESIONES REGULARES

Una expresión regular (abreviada como regex [...]) es una secuencia de caracteres que especifica un patrón de búsqueda en un texto [...] implementado por algoritmos de búsqueda de cadenas de texto para "encontrar" o "encontrar y reemplazar" o para validare entradas.

Los pasos para realizar las búsquedas son:
1. Importar la librería re.
2. crear un objeto Regex por medio de re.compile().
3. Pasar la cadena de texto que se quiere buscar al objeto search(). Esto regresa un objeto tipo Match().
4. Llamar al método group() del objeto Match.

Por ejemplo, busquemos números de teléfonos celulares de la forma ###-###-####

```python
#1 Se importa la librería re
import re
```

La r antes de las comillas significa "expresión cruda" (raw expression). LA expresión cruda se usa para que reconozca como caracteres símbolos que podrían ser ambiguos como */?, etc.

```python
#2 se crea el objeto num_tel 
num_tel = re.compile(r'(\w+\s|\s\w+)')
#dir(num_tel)
```

```python
#Cadena de caracteres a pasar en la búsqueda
mi_num = 'Mi número es 300-666-7777. Llámame'

#3 Se pasa el método search() al objeto num_tel creado

num_extr = num_tel.findall(mi_num)

num_extr
```

```python
#dir(num_extr)
```

```python
#4. Se llama el método group del objeto num_extr
num_extr.group()
```

En este caso, el patrón de búsqueda que se pasó al método compile fue '\d\d\d-\d\d\d-\d\d\d\d', donde '\d' significa 'BUSQUE CUALQUIER DÍGITO'.

```python
#¿Qué pasaría si en lugar de \d se hubiera puesto \D'?
```

##AGRUPAMIENTO CON PARÉNTESIS

```python
num = re.compile(r'(\d\d\d)-(\d\d\d)-(\d\d\d\d)')
grupos = num.search(mi_num)
```

```python
grupos.group()
```

```python
grupos.group(0)
```

```python
grupos.groups(4)
```
