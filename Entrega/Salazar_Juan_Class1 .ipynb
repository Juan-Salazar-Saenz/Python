Ejercicio 1. LISTA DE NÚMEROS PRIMOS

Un número natural es primo si es divisible por 1 y por el número mismo solamente. Por
ejemplo, 7 es primo, ya que es divisible entre 1 y 7. Por otra parte, 4 no es primo, ya que es
divisible entre 1, 2 y 4. En las siguientes rutinas, se les piden que las hagan usando el
método .isprime()

a. Escriba una rutina que genere los números primos hasta un número N (N siendo
entero positivo).

b. Escriba una rutina que genera una lista con los N primeros números primos.

Respuesta

a. Escriba una rutina que genere los números primos hasta un número N (N siendo entero positivo).

```python
def es_primo(num):
  for n in range(2, num):
    if num % n == 0:            
     return False
  return True

def imprimir(encontrarPrimos):
  for i in range(encontrarPrimos + 1):
    if i > 1:
      esprimo = es_primo(i)
      if esprimo:
        print(i)

while True:
  try:
    x = int(input("Maximo para buscar primos : "))
    print("Los numeros primos son :")
    imprimir(x)
  except:
    break
```

b. Escriba una rutina que genera una lista con los N primeros números primos.

```python
def es_primo(num):
  for n in range(2, num):
    if num % n == 0:            
     return False
  return True

def imprimir2(encontrarPrimos):
  buscar = True
  i = 2
  encontrados = 0
  while buscar:
    esprimo = es_primo(i)
    if esprimo:
      print(i)
      encontrados += 1
      if encontrados == encontrarPrimos:
        buscar = False
    i += 1

while True:
  try:
    x = int(input("Maximo primos a buscar : "))
    imprimir2(x)
  except:
    break
```

Ejercicio 2. SERIE DE FIBONACCI

La serie de Fibonacci se construye a partir de ubicar 1,1 en una lista. De ahí en adelante,
cada número nuevo es el resultado de la suma de los dos números inmediatamente
anteriores. Por ejemplo, la serie continúa como 1, 1, 2; luego, 1, 1, 2, 3; 1, 1, 2, 3, 5; etc.
Genere un código que genere los N primeros números de Fibonacci, siendo N un número
entero positivo cualquiera.

```python
from math import sqrt

def F(n):
  return ((1+sqrt(5))**n-(1-sqrt(5))**n)/(2**n*sqrt(5))

def Fibonacci(endNumber):
  n = 0
  cur = F(n)
  while n <= endNumber:
    print("{:.0f}".format(cur))
    n += 1
    cur = F(n)

while True:
  try:
    x = int(input("Digite el numero de Fibonacci "))
    if x > 0:
      Fibonacci(x-1)
  except:
    break
```

Ejercicio 3. NÚMERO DE VOCALES DE UN TEXTO

Genere un código que devuelva un diccionario que indique cuántas veces se repite cada
vocal en un texto.

```python
def imprimir(entrada):
  entrada = entrada.lower()
  vocales = "a e i o u"
  vectorVocales = vocales.split(' ')
  frecuenciaVocales = []
  for w in vectorVocales:
      frecuenciaVocales.append(entrada.count(w))

  print("Frecuencia vocales\n" + str(list(zip(vectorVocales,     frecuenciaVocales))))

while True:
  try:
    x = input("Digite el texto para buscar las vocales: ")
    imprimir(x)
  except:
    break
```

Ejercicio 4. PALABRAS MÁS REPETIDAS EN UN TEXTO

En la carpeta se adjuntan dos archivos .txt. En stopwords.txt se incluyen palabras del tipo
aunque, con, solamente, etc. En servicio_geologico.txt se incluye una lectura sobre el
servicio geológico.

a. Genere un diccionario que diga cuáles son las palabras más repetidas con sus
respectivas frecuencias.

b. Genere un segundo diccionario con las palabras del texto que sean stopwords.

c. Repita el ejercicio a, pero esta vez omita las palabras que sean stopwords.

```python
def imprimir(entrada):
  vector = entrada.split(' ')
  frecuenciaPalabras = []
  for w in vector:
    frecuenciaPalabras.append(vector.count(w))
    
  print("Cadena\n" + entrada +"\n")
  print("Lista\n" + str(vector) + "\n")
  print("Frecuencias\n" + str(frecuenciaPalabras) + "\n")
  print("Pares\n" + str(list(zip(vector, frecuenciaPalabras))))
  
while True:
  try:
    x = input("Digite el texto : ")
    imprimir(x)
  except:
    break
```

Bonus : Serie pascal

```python
def triangle_pascal(rows):
  if rows == 0:
    return []
  elif rows == 1:
    return [[1]]
  else:
    new_row = [1]
    result = triangle_pascal(rows - 1)
    last_row = result[-1]
    for i in range(len(last_row) - 1):
      new_row.append(last_row[i] + last_row[i + 1])
    new_row += [1]
    result.append(new_row)
  return result

def imprimir(tamano):
  triangulo = triangle_pascal(tamano)
  ultimaLinea = triangulo.pop()
  triangulo.append(ultimaLinea)

  p = ""
  for j in ultimaLinea:
    p = p + ' ' + str(j)
  ultimo = len(p)

  for x in triangulo:
    p = ""
    for j in x:
      p = p + ' ' + str(j)

    t = p.center(ultimo)
    print(t)

while True:
  try:
    x = int(input("Digite el numero de pascal "))
    imprimir(x)
  except:
    break
```
