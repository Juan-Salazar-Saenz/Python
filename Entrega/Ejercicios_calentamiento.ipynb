Escriba una rutina que reciba dos ángulos de un triángulo y regrese el tercer valor. Asuma que el usuario no cometerá errores.

```python
Angulo1 = int(input( "Primer angulo : ") )
Angulo2 = int(input( "Primer angulo : ") )

if ((Angulo1 + Angulo2) > 180):
  print("Parametros invalidos")
else:
  print("El tercer angulo es :" , (180 - (Angulo1 + Angulo2)))
```

Escriba una rutina que acepte los nombres y los apellidos de una persona y los imprima de la forma: 



*   Entrada: Donny Barcenas
*   Salida: Barcenas, Donny



```python
Nombre = input( "Escriba su nomnbre : ") 
todo = Nombre.split(' ')
formato = ""
for w in todo :
  formato = formato  + w + ','
print(formato)
```
